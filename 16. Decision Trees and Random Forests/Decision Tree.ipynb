### Decision Tree
- Simple Tree like structure, model makes a decision at every node
- Useful in simple tasks
- one of the most popular algorithm
- Easy explainability, easy to show how a decision process works!
- Its a np hard problem
- Its a Exponential Time complexity.

### Why decision trees are popular ?
- Easy to interpret and present
- Well defined Logic, mimic human level thought
- Random Forests, Ensembles of decision trees are more powerful classifier
- Feature values are preferred to be categorical. If the values are continuous then they are discretized prior to building the model.

### Build Decision Trees
**Two Common Algorithms**

- CART (Classification and Regression Trees) $\rightarrow$ uses Gini Index(Classification) as metric.
- ID3 (Iterative Dichotomiser 3) $\rightarrow$  uses Entropy and Information gain as metrics.

**How to Split datas? Which feature to choose to split on?**
- We use Greedy Approach
- At a step/level we will go through all features and find out which feature to split upon 
- Like if we split on feature1 what gain/accuracy will I get, then check on other features.
- Base Case 1. Start the node, if node is pure(all the data here belongs A nothing belongs to B) then output only class that is there.
   Base Case 2.: If no feature is left to split upon.

- If both base condition fails, then find best feature to split upon
- Recursively call on these splits.

**Objective Metric** tells how good or bad is the decision tree.

### Entropy $\rightarrow$ measures randomness of the system
Randomness :-
H(S) $\rightarrow$ H = Entropy and S = System
- $H(S) = -\sum P_{c}Log_{2} P_{c} $
- $P_{c} \rightarrow$ Probability of class c
- Example:
    - 3R, 2G, 1W
    - H(S) = -(3/2 log 3/6 + 2/6 log 2/6 + 1/6 log 1/6)

### Information Gain
- Its a standard way of deciding which feature to split on.
- Information Gain(IG)
- S = Set, A = Attribute on which split is done, Sv = New Set 
- Feature we choose should lead to maximum decrease in the IG or max Decrease in Entropy.
- $\textbf{Higher IG lower the Impurity}$
---------------------------------------------------------------------------------------


**Root Node Will Be IG(S, Outlook) = 0.247 Because It is give maximum reduction in ENTROPY**

### ENTROPY DETAILED EXPLANATION
- Let say we have a dataset with features as f1, f2, f3 .. and binary(Y/N) output column.
- While contructing decision tree there is an algorithm called ID3 algorithm, which says that while contruction D tree first step is to select correct/right attribute(feature) to construct a D Tree.
- Which Feature to choose then ?, that mean we need to use Entropy technique.
- Entropy measures purity of the split.
- Perfect impure is When we have something like 50% of yes ans 50% noes, will give Entropy of 1 always.
- Entropy values ranges from 0 to 1 
- Entropy is calculated for single node
- Example : 9Yes and 3 Nos
        - Entropy -> H(S) = P(+ve)log(P(-ve)) - P(-ve) log(P(-ve))
                    here = 9/12 log(9/12) - 3/12 log(3/12)
- Information Gain combines all Entropy together

### Working with the sklearn 

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
x_train,x_test,y_train,y_test= train_test_split(iris.data, iris.target, random_state = 1)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train, y_train_pred) # 0 shows 0 erros that means it could be overfitting
```

```python
confusion_matrix(y_test,y_test_pred) # 1 shows 1 error while trainset shows 0 error
```
