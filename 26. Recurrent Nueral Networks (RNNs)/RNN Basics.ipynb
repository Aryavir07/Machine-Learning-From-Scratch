<h1><center><strong>Sequence Models (RNNs/LSTM)</strong></center></h1>


- Idea of Sequence model is to process sentence words by words.
- We iterate over the sentence from left to right as human reads it.
- We will maintain a STATE, its like what we read/processed till now.
    - Models like MLP or Feed Forward Network and CNN's they not remember thier past processings.
    - Like if any CNN model, processed any image then again we work on the image it will run from strach over it as it doesnot remember its past.
- Use of sequence model:
    - Lets say we have a sentence and we want to rate it on scale of 5
        - "This is a great movie". 5/5 (say)
        - "This is a great movie, but ending was boring" :- 4/5 rating
        - To do this kind of work we use RNNs
        
- Recurrent mean applying some function on every word.
- RNNs it has a State Vector and RNN cell.
    - RNN cell takes words as input, processes it and then return some output.
    - State vector, it stores information about the words that it has seen.
    
- When an RNN cell takes input(words of sentence) to it produces an output and activation/State for next word..
- State vector will maintain what you have seen so far.
