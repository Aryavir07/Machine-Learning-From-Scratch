## Linear Regression Intro
 $y = mx + c + noise$


$m =  \theta_{0}, c =  \theta_{1}$
<br>
$\theta = $$\begin{bmatrix} \theta_{0} \\ \theta_{1}\end{bmatrix}$
### Goal is to find out best value of $\theta$
- Step 1: Initialize $\theta$ with random value
- Step 2: Measure how good is our $\theta$ in terms of accuracy, such that it gives best fit line with less error
- Step 3: How to measure accuracy ? : By measuring Error
- Step 3: Update Theta such that $J{(\theta)}$ or Error is less
-   Lesser Error, Better Accuracy
- $Error_{i} = | Y^{(i)}_{Actual} - Y^{(i)}_{Predicted} |$
- This for one example
- Finding total Error over all examples
- Total Error $=\sum_{i=1}^{m}|Y^{(i)}_{pred} - Y^{(i)}_{actual} |$
- m is total number of example 
-  Average Error $=\frac{1}{m}\sum_{i=1}^{m}|Y^{(i)}_{Actual} - Y^{(i)}_{Pred} |$
**Average Error above is not Differential Function its Graph is like V**
- To fix it, we make graph like U by squaring it Which is known as Mean Squared Error (MSE)
- MSE $=\frac{1}{m}\sum_{i=1}^{m}|Y^{(i)}_{Actual} - Y^{(i)}_{Pred}|^{2}$

## Gradient Descent
- It is method to find local minima of any function
- Gradient Descent is Used to Minimize the Cost/Error Function
- Start with any random value of $x = 0$ (let)
- Take small step toward minima $(0)$ such that your step size is getting smaller
- Direction is toward downhill
<br>
$y = f(x)$
<br>
- Learning rate is always taken small
- Let the there be a function $y = f(x)$
<br>
- **Goal** to find value of x which minimizes $f(x)$
- update $x$ as $x=x-learningRate* \frac{\partial f(x)}{\partial x}$

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

**Goal** Given A Function f(x), we want to find the value of x that minimizes f.

### Visualization

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('F(x)')
plt.show()
```

```python
x = 0
lr = 0.1
error = []
# Take 50 Steps in Downhill direction
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5) # derivative of above equation
    x = x - lr*grad 
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
# Plot the values of error
plt.plot(error)
# Shows error decrease upto 20 steps then become almost saturated
plt.show()
```

- Steps we do for linear regression single valued
- 1. $\theta = init()$
- 2. $J(\theta)$  $=\frac{1}{m}\sum_{i=1}^{m}|Y^{(i)}_{Actual} - Y^{(i)}_{Pred}|^{2}$
    - $\theta$ minimizes error function $J(\theta)$
    - $\theta = [\theta_{0},\theta_{1}] 1$ 
- 3. update $\theta$ to minimize loss using gradient descent 

$Y^{i}_{pred} = h_{\theta}(X) = \theta_{0} + \theta_{1}*x^{(i)}$
- $J(\theta)$ will be bowl Like structure in 3D

### Derivation 
- $J(\theta)$  $=\frac{1}{m}\sum_{i=1}^{m}|Y^{(i)}_{Pred} - Y^{(i)}_{Actual}|^{2}$
- $J(\theta)$  $=\frac{1}{m}\sum_{i=1}^{m}|[\theta_{0} + \theta_{1}x^(i)] - Y^{(i)}_{Actual}|^{2}$
<br>
$\theta_{0} = \theta_{0} - learningRate* \frac{\partial J(\theta)}{\partial \theta_{0}} \dots\dots (1)$
<br>
$\theta_{1} = \theta_{1} - learningRate* \frac{\partial J(\theta)}{\partial \theta_{1}} \dots\dots (2)$
- On derivation of $J(\theta)$ wrt $\theta_{0}, \theta_{1}$

## Linear Regression

```python
def hypothesis(x,theta):
    # theta = [theta0, theta1]
    y_pred = theta[0] + theta[1]*x
    return y_pred

def gradient(X,Y, theta):
    # Y is complete dataset
    m = X.shape[0] # 3750 
    grad = np.zeros((2,)) #array([0,0])
    
    for i in range(m):
        y_pred = hypothesis(X[i], theta)
        y = Y[i]
        grad[0] += (y_pred - y) 
        grad[1] += (y_pred - y)*X[i]
        
    return grad/m

def error(X,Y,theta):
    m = X.shape[0]
    total_error = 0.0
    for i in range(m):
        y_pred = hypothesis(X[i], theta)
        total_error += (y_pred - Y[i])**2
    return total_error/m
        
def gradientDescent(X,Y,max_step = 100, lr = 0.1):
    theta = np.zeros((2,))
    error_list = []
    theta_list = []
    for i in range(max_step):
        # Compute grad
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        # update theta, for that we need gradient
        theta[0] = theta[0] - lr*grad[0]
        theta[1] = theta[1] - lr*grad[1]
        theta_list.append((theta[0], theta[1]))
        # Function to compute J(theta)
    return theta, error_list, theta_list
```

```python
theta, error_list, theta_list= gradientDescent(X,y)
```

## Section - 4 Computing Score
**Score : R2 (R-Squared) or Coefficient of determination**


- R2 score = 1 - summation from 1 to m (y(i) pred- y(i)actual)^2/summation from 1 to m (y(i)- y(i)avg)^2
- R2 score is used find accuracy
- R2 Score ranges from 0 to 1 
- so .97 = .97*100 = 97% accuracy
- R2(R-squared) AKA Coefficient of Determination

```python
def r2_score(Y,Y_):
    # Instead of Loop, np.sum is recommended as it is fast
    num = np.sum((Y-Y_)**2)
    denom = np.sum((Y- Y.mean())**2)
    score = (1- num/denom)
    return score*100
```
