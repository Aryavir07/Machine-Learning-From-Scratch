### Working On Non-Linearly Saperable Datasets

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons
```

```python
X,Y = make_moons(n_samples = 200, noise = 0.10)
```

```python
plt.scatter(X[:,0],X[:,1],c = Y)
plt.show()
```

```python
Y.shape, X.shape
```

```python
Z,K = make_circles(n_samples = 200, noise = 0.10)
```

```python
Z.shape
```

```python
K.shape
```

```python
plt.scatter(Z[:,0],Z[:,1],c = K)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 2)
km.fit(X)
```

```python
centers = km.cluster_centers_
label = km.labels_
```

```python
plt.scatter(X[:,0],X[:,1], c = label)
plt.scatter(centers[:,0], centers[:,1],marker = '*', color = 'red')
plt.show()
# KMeans fails here with the complex datas
```

### To fix this problem of KMean we use DBSCAN Algorithm
- Density based Spatial Clustering of Application with noise
- Based upon on the idea, that a cluster is a high density area surrounded by low density region
- Starts by exploring a small area, if density is 'good enough' it considered as part of cluster, and explores neighbours and so on, so it groups region based upon density

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db =DBSCAN(eps=0.21,min_samples=5)
```

```python
db.fit(X)
```

```python
Ypred = db.fit_predict(X)
```

```python
print(Ypred)
```

```python
plt.scatter(X[:,0],X[:,1], c = Ypred)
plt.show()
```
