## Normalization and Standardization 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline 
```

- What is Normalization or Standardization ?
- - To make mean of data as 0 and standard deviation as 1
- Why normalization ?
- - Lets consider a dataset about a car whose weight loading capacity is in tones and time is given as Centimeter/hr, this cm/hr on plotting scatter plot it will scatter a lot. To bring those values into a given range we normalize the data

```python
from sklearn.datasets import load_digits
```

```python
dfx = pd.read_csv('./datasets/linearX.csv')
dfy = pd.read_csv('./datasets/linearY.csv')
```

```python
x = dfx.values
y = dfy.values
```

```python
plt.scatter(x,y)
```

```python
x1 = (x - x.mean())/x.std()
y2 = (y - y.mean())/y.std()
```

```python
plt.scatter(x,y,label = 'raw data')
plt.scatter(x1,y,color='red',label = 'normalized data')
plt.legend()
```
