```python
import cv2
```

```python
img = cv2.imread('download.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img) # RGB 3 channels we read img using opencv, openCv treats
# image BGR not as RGB so Red colors are swapped with Blue 
```

```python
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # for swapping color channels 
```

```python
plt.imshow(newImg) # back to original color
```
