### BROADCASTING

- You can add a scalar to a vector, and numpy will add it to each element in the vector
$x + a = x_{i} + a$
- Similarly you can add a vector to a matrix, and numpy will add the vector to each column of the matrix

```python
x = np.array([1,2,3,4])
print(x)
x = x + 5
print(x)
```

```python
x = np.array([[1,2,3,4],[5,6,7,8]])
x
```

```python
x = x + 5
x
```

### Matrix Multiplication

This is perhaps one operations that you would use quite frequently in any ML/DL model. you should remember a few things about multiplication

- $C$ = $AB$ is only defined when the second dimension of A mathes the first dimension of $B$
- Further, If A is of slope (m,n) and $B$ of shape (n,p), then $C$ is of shape (m,p)
- This operation is concretely defined as $C_{i,j}= \sum_{k}A_{i,k}B_{k,j}$
  - $C_{i,j}$ is computed by taking the dot product of i-th row of $A$ with j-th column of $B$
- A more useful method to think of matrix multiplication is as __Linear Combination Of Columns__ of $A$ weighted by columns entries of $B$

## Norms


```python
x = np.array([3,4])
lp2 = np.linalg.norm(x)
lp2
```

## Determinants

```python
A = np.array([[1,2],[3,4]]) # 1 x 4 - 2 x 3 =  4 - 6 = -2
np.linalg.det(A)
```

## Inverse

```python
Ainv = np.linalg.inv(A)
Ainv
# A dot Ainv = 1 
```

```python
np.dot(A,Ainv)
```

```python
pinv = np.linalg.pinv(A)
print(pinv)
```

```python
np.dot(A,pinv)
```

```python
A = np.array([[6,8],[3,4]])
pinv = np.linalg.pinv(A)
print(pinv)
```
