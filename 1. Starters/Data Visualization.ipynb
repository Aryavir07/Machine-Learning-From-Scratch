### DATA VISUALIZATION CHALLENGE
Given a Movie Dataset, your task is to make a visualization that:
- Plot the length of movie title names on X-axis
- Frequently counts on Y-axis i.e Number of Movies having 'x' character in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read A CSV File
df = pd.read_csv("./Datasets/movie_metadata.csv")
df.head(10)
```

```python
df.columns
```

```python
titles = list(df.get('movie_title'))
titles[:5]
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
freq_titles
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length Of Movie Title")
plt.xlabel("Freq count")
plt.title("Movie Data Visualization Problem")
plt.show()
```
