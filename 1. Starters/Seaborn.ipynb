```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data= tips)
```

```python
sns.countplot(x = 'time', data = tips)
```

```python
sns.boxplot(x= 'day',y= 'total_bill', data = tips) # show min and max values
```

```python
sns.boxplot(x= 'day',y= 'total_bill', data = tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex',split = True)
```

```python
sns.distplot(tips['total_bill'], bins = 30)
```

```python
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x ='total_bill' , y = 'tip' , data = tips)
# this plot shows that
# if total_bill is 10 then customer gives 1 as tips similary for 50 its around 9 
# that shows it is positive line increasing positively
```

```python
sns.jointplot(x ='total_bill' , y = 'tip' , data = tips, color ='r', kind = 'hex')
```
