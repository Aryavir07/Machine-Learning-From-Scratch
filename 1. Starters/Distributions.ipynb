### Multivariate Normal Distribution

```python
# cov - covariance matrix
# cov = np.array([[1,0],[0,1]]
#                [[xx,xy],[yx,yy]]
```

```python
mean = np.array([0.0,0.0]) #let
cov = np.array([[1,0],[0,1]]) # considering Xaxis = height, Yaxis = Color
# then Xaxis and Yaxis have no correlation
dist = np.random.multivariate_normal(mean, cov, 500)
```

```python
dist
```

```python
dist.shape 
```

```python
plt.scatter(dist[:,0],dist[:,1])
```

```python
mean = np.array([0,0]) #let
cov = np.array([[1,1],[1,1]]) # mean highly corelated values
dist = np.random.multivariate_normal(mean, cov, 500)
plt.scatter(dist[:,0],dist[:,1])
```

```python
mean = np.array([0,0]) #let
cov = np.array([[1,.8],[.8,1]])
dist = np.random.multivariate_normal(mean, cov, 500)
plt.scatter(dist[:,0],dist[:,1])
```

```python
mean = np.array([0,0]) #let
cov = np.array([[1,.8],[.8,1]])

mean2 = np.array([5.0,6.0])
cov2 = np.array([[1.3,.2],[.3,1.1]])
dist2 = np.random.multivariate_normal(mean2, cov2, 500)
dist = np.random.multivariate_normal(mean, cov, 500)
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
plt.show()
```
